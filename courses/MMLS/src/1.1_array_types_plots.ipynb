You can use libraries to extend python with functionality.

```python
# The "import" statement pulls in the packages.
# I do it differently from the professor, which imports everything with
# from matplotmib.pylab import *

# This is cleaner, and avoids problems in more complex programs.
# The functions are called as `pl.<fun>()` instead of just `<fun>()`
import matplotlib.pylab as pl
import numpy as np
```

```python
# Example: vectors are extended lists
a = np.array([1,2,3,4,5])

print(a)

# Calculations on vectors are... vectorized, so they are
# applied on each value in order.
print(a * 2)
print(a ** 2.5 / (a + 3))
```

```python
# You can select stuff from vectors with the [] operator
# As in <var>[<position>]
# Python is zero-indexed, so the first item is in positio 0
a[0]
```

Numpy has stronger (and different) types than Python. E.g. there is a `np.int32`, instead of just `int`. This is because the math behind the scenes is handled differently in numpy than core Python.

```python
print(np.int32(2))
```

```python
a = np.int32(20)
print(type(a))

a **= 31
print(a)
print(type(a)) # See the conversion?
```

```python
# We can generate numbers easily, e.g. in linear space
x = np.linspace(-2, 5, 50) # from -2, to 5, giving 50 numbers
print(x)
```

```python
# As math stuff is vectorized over numpy vector, we can do this
fx = x**2-3*x+2
print(fx)
```

```python
pl.plot(x, fx, ":g", label = "x**2-3*x+2") # The third parameter controls the line aestetic
pl.legend()
pl.grid() # Add a grid
```

```python
# We can also add more lines
fx2 = -0.5*x**2 * 10*x -10

pl.plot(x, fx, ":g", label = "x**2-3*x+2")
pl.plot(x, fx2, "b", label = "-0.5*x**2 * 10*x -10")
pl.legend()
pl.grid() # Add a grid
```

```python
# You can also save and load data.
np.savetxt("example.txt", fx)

fx_copy = np.loadtxt("example.txt")

# "all" takes a list (like that made by fx == fx_copy, of TRUE, TRUE, TRUE...)
# and tests if all inputs are TRUE
assert all(fx == fx_copy)
```
