More derivative exercises.

```python
import matplotlib.pylab as pl
import sympy as sp
```

```python
space = pl.linspace(-10, 10, 2000) # A generic range

def fn1(x):
    return pl.sin(x)

pl.plot(space, fn1(space))
```

```python
# We can now calculate the aproximate derivative.
# Slightly differenty than last time, let's define a delta:
delta = 1.5

# We can now try and calculate the derivative given the delta at some
# point c
print(f"The point differental at point {2} is {(fn1(2+delta) - fn1(2))/delta}")

# We can now calculate this for all points...
def approx_derivative(x, function, delta):
    return ( function(x+delta) - function(x) ) / delta

print(approx_derivative(space, fn1, delta))
```

```python
# Now some plotting
pl.plot(space, fn1(space), label="original")
pl.plot(space, approx_derivative(space, fn1, 1.5), label = "approx")
# As we crank delta to be smaller, the approximation gets better and better
pl.plot(space, pl.cos(space), label="real")
pl.legend
```

```python
pl.plot(space, fn1(space), label="original")
pl.plot(space, approx_derivative(space, fn1, 0.5), label = "approx")
# As we crank delta to be smaller, the approximation gets better and better
pl.plot(space, pl.cos(space), label="real")
pl.legend
```

```python
pl.plot(space, fn1(space), label="original")
pl.plot(space, approx_derivative(space, fn1, 0.5), label = "approx")
# As we crank delta to be smaller, the approximation gets better and better
pl.plot(space, pl.cos(space), label="real")
pl.legend
```

See how the function "shifts" to the left as `delta` gets larger? This is because we are calculating the slope of the function between the point and a point just forwards (+delta), so we're not actually calculating the slope *around* the point of interest. To do this, we would need to look at `f(x+(delta/2)) - f(x-(delta/2))`.

## Second order derivatives

The second order derivatives are useful to check for (local) minimal/maximal points.

If the second order derivative has a point that is equal to zero, and in which it is of opposing signs on the left and right of the point, then that is a (local) minimum/maximum.
If the signs are equal, it means that the function keeps on growing/decreasing after the apparent minimum or maximum.

#### De L'Hopital's rule
If the limits of two functions are 0 or $\pm \infty$, we can say that:

$$
\lim_{x \to c}\frac{f(x)}{g(x)} = \lim_{x \to c}\frac{f'(x)}{g'(x)}
$$
